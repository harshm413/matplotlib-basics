```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
y2 = [1, 4, 2, 9, 5]
```

```python
fig,ax=plt.subplots(2,1)
ax[0].plot(x,y)
ax[0].set_xlabel('x-axis')
ax[0].set_ylabel('y-axis')
ax[0].set_title('chart 1')
ax[1].scatter(x, y2)
ax[1].set_xlabel('X-Axis')
ax[1].set_ylabel('Y-Axis')
ax[1].set_title('Chart 2')
plt.tight_layout()
```
