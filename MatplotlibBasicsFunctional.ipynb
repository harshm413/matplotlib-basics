```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
y2 = [1, 4, 2, 9, 5]
```

```python
plt.subplot(2,1,1)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('chart 1')
plt.plot(x,y)
plt.subplot(2,1,2)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('chart 2')
plt.scatter(x,y2)
```
